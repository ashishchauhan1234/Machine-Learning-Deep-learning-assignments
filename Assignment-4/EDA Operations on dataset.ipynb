```python
plt.figure(figsize = (15,5))
order_sepal_length = data.groupby(['species']).mean().sort_values('sepal_length', ascending = False).index
sns.barplot(data['sepal_length'],data['species'],order = order_sepal_length )
```

```python
plt.figure(figsize = (15,5))
order_sepal_width = data.groupby(['species']).mean().sort_values('sepal_width', ascending = False).index
sns.barplot(data['sepal_width'],data['species'],order = order_sepal_width)
```

```python
plt.figure(figsize = (15,5))
order_petal_length = data.groupby(['species']).mean().sort_values('petal_length', ascending = False).index
sns.barplot(data['petal_length'],data['species'],order = order_petal_length)
```

```python
plt.figure(figsize = (15,5))
order_petal_width = data.groupby(['species']).mean().sort_values('petal_width', ascending = False).index
sns.barplot(data['petal_width'],data['species'],order = order_petal_width)
```

```python
fig, subplot_axis = plt.subplots( nrows= 1, ncols= 4, sharey = True,figsize = (20,5))
subplot_axis[0].set_title("sepal length vs species")
sns.barplot(data['sepal_length'],data['species'],order = order_sepal_length,ax =subplot_axis[0] )
subplot_axis[1].set_title("sepal width vs species")
sns.barplot(data['sepal_width'],data['species'],order = order_sepal_width, ax = subplot_axis[1])
subplot_axis[2].set_title("petal length vs species")
sns.barplot(data['petal_length'],data['species'],order = order_petal_length,ax = subplot_axis[2])
subplot_axis[3].set_title("petal width vs species")
sns.barplot(data['petal_width'],data['species'],order = order_petal_width, ax = subplot_axis[3])
plt.show()
```

```python
from sklearn.preprocessing import OrdinalEncoder
oc = OrdinalEncoder()
cols = ['species']
label_data = data['species'].unique()
data[cols] = oc.fit_transform(data[cols])
```

```python
data.head()
```

```python
data['species'].unique()
```

```python
pie_data = data.groupby(['species']).mean()
pie_data
```

```python
fig, subplot_axis = plt.subplots(nrows= 2, ncols= 2, figsize = (15,10))
subplot_axis[0,0].set_title("Pie chart of Sepal length column")
subplot_axis[0,0].pie(pie_data['sepal_length'], labels = label_data, autopct = "%.1f%%")
subplot_axis[0,1].set_title("Pie chart of Sepal width column")
subplot_axis[0,1].pie(pie_data['petal_width'], labels = label_data, autopct = "%.1f%%")
subplot_axis[1,0].set_title("Pie chart of petal length column")
subplot_axis[1,0].pie(pie_data['petal_length'], labels = label_data, autopct = "%.1f%%")
subplot_axis[1,1].set_title("Pie chart of petal width column")
subplot_axis[1,1].pie(pie_data['petal_width'], labels = label_data, autopct = "%.1f%%")
plt.show()
```
